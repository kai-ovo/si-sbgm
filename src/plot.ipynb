```python
from utils import plot_setting  
plot_setting()
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

read data


```python
d = 2
data2 = np.load(f'./data/gmm{10}d.npy',allow_pickle='TRUE').item()
si_result2 = np.load(f'./result/gmm{10}d.npy',allow_pickle='TRUE').item()
sbgm_result2 = np.load(f'./result/sbgm-gmm{10}d.npy',allow_pickle='TRUE').item()
```

```python
si_mean2 = si_result2['mean']
si_cov2 = si_result2['cov']
si_wass2 = si_result2['w2']

sbgm_mean2 = sbgm_result2['mean']
sbgm_cov2 = sbgm_result2['cov']
sbgm_wass2 = sbgm_result2['w2']
```

```python
si_steps    = [
    10,20,40,80,160,320,640,1280
]
sbgm_steps    = [
    10,20,40,80,160
]
plt.figure(figsize=(6,4))
plt.semilogx(si_steps, si_mean2, '--x', markersize= 10,label='SI')
plt.semilogx(sbgm_steps, sbgm_mean2, '--x', markersize= 10,label='SBGM')
plt.title("Error in Mean", fontsize=20)
plt.legend(prop={'size':12})
```

```python
plt.figure(figsize=(6,4))
plt.semilogx(si_steps, si_cov2, '--x', markersize= 10,label='SI')
plt.semilogx(sbgm_steps, sbgm_cov2, '--x', markersize= 10,label='SBGM')
plt.title("Error in Covariance", fontsize=20)
plt.legend(prop={'size':12})
```

```python
plt.figure(figsize=(6,4))
plt.semilogx(si_steps, si_wass2, '--x', markersize= 10,label='SI')
plt.semilogx(sbgm_steps, sbgm_wass2, '--x', markersize= 10,label='SBGM')
plt.title("Entropy-Regularized W-2 Distance", fontsize=20)
plt.legend(prop={'size':12})
```
